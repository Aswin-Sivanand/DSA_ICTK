```python
corr_matrix = wine_df.corr(numeric_only = True)
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
#k means


from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,max_iter=300,random_state = 0, n_init=20)
  kmeans.fit(wine_df)
  css = kmeans.inertia_
  print('Clusters = ',i,'wcss = ',css)
  wcss.append(css)
```

```python
#Elbow

plt.figure(figsize=(8,6))
plt.plot(range(1,11),wcss,linewidth=2,markersize=8, color="green")
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,11,1))
plt.grid()
plt.show()
```

```python
#Taking number of clusters as 3

km = KMeans(n_clusters=3,n_init=20)
cluster_labels = km.fit_predict(wine_df)
cluster_labels
```

```python
#Agglomerative

from seaborn.matrix import dendrogram
import scipy.cluster.hierarchy as sch
dendrogram  = sch.dendrogram(sch.linkage(wine_df,method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Data points')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ahc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
ahc
```

```python
y_ahc = ahc.fit_predict(wine_df)
y_ahc
```

```python
plt.scatter(wine_df["Alcohol"],wine_df["Proline"],c=cluster_labels)
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.title('Alcohol vs Proline')
plt.show()
```

```python
plt.scatter(wine_df['Alcohol'], wine_df['Proline'], c=ahc.labels_)
plt.title('Agglomerative Clustering')
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_ahc = silhouette_score(wine_df,y_ahc)
sil_ahc
```

```python
#DBScan

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import make_blobs
```

```python
neighbors = NearestNeighbors(n_neighbors=12)
neighbors_fit = neighbors.fit(wine_df)
distances, indices = neighbors_fit.kneighbors(wine_df)

# Plot the k-distance graph
distances = np.sort(distances, axis=0)
distances = distances[:, 1]
plt.plot(distances)
plt.xlabel('Data Points')
plt.ylabel('Epsilon value')
plt.title('K-Distance Graph for DBScan')
plt.show()
```

```python
#from graph we get eps as 35
db = DBSCAN(eps=35, min_samples=12).fit(wine_df)
dmodel = db.fit(wine_df)
labels = db.fit_predict(wine_df)
labels
```

```python
# number of Clusters

smple_cores = np.zeros_like(labels,dtype=bool)
smple_cores[db.core_sample_indices_]==True
n_clusters = len(set(labels))-(1 if -1 in labels else 0)
print("Number of Clusters = ",n_clusters)
```

```python
# Noise points
n_noise = list(labels).count(-1)
print("Number of Noise Points = ",n_noise)
```

```python
# Silhoutte Score
print('Silhoutte Score is',silhouette_score(wine_df,labels))
```

```python
# Visualizing the clusters
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
wine_df_pca = pca.fit_transform(wine_df)


plt.scatter(wine_df_pca[:, 0], wine_df_pca[:, 1], c= labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```
