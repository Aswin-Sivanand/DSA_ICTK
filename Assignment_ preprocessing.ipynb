```python
#Check and handle outliers in at least 3 columns in the dataset
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

num_columns=['Parch', 'Age', 'Fare']
#boxplot
for i in num_columns:
    plt.figure()
    plt.boxplot(titanic_df[i])
    plt.title(i)
```

```python
#handling outliers in column parch
P1 = np.percentile(titanic_df["Parch"],25,method="midpoint")
P2 = np.percentile(titanic_df["Parch"],50,method="midpoint")
P3 = np.percentile(titanic_df["Parch"],75,method="midpoint")
P1,P2,P3
```

```python
IPR = P3 -P1


low_lim = P1 - 1.5*IPR
up_lim = P3 + 1.5*IPR
print("min is", low_lim)
print("max is", up_lim)
```

```python
outlier = []
for x in titanic_df['Parch']:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
        ind1 = titanic_df["Parch"]>up_lim
titanic_df.loc[ind1].index
```
