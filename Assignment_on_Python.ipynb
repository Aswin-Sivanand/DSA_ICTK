```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the dataset to python

data = pd.read_excel("iris.xls")
```

```python
#Display the columns in the dataset.

data.columns
```

```python
# Calculate the mean of each column of the dataset

Mean_SL = data["SL"].mean()
Mean_SW = data["SW"].mean()
Mean_PL = data["PL"].mean()
Mean_PW = data["PW"].mean()
print(Mean_SL,Mean_SW,Mean_PL,Mean_PW)
```

```python
#Check for the null values present in the dataset

data.isnull().sum()
```

```python
print(data.info())
```

```python
#Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# scatterplot
colors = ['red', 'Yellow', 'Violet']
classification = ['Iris-virginica','Iris-versicolor','Iris-setosa']
```

```python
data['Classification'].value_counts()
```

```python
sns.scatterplot(x='SL', y='SW',
                hue='Classification', data=data, )
```

```python
#Histogram, Histplot

plot = sns.FacetGrid(data, hue="Classification")
plot.map(sns.histplot, "SL").add_legend()
```

```python
for i in range(3):
    x = data[data['Classification'] == classification[i]]
    plt.scatter(x['PL'], x['PW'], c = colors[i], label=classification[i])
plt.xlabel("PL")
plt.ylabel("PW")
plt.legend()
```
